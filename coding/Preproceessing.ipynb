Visualizing the Data


```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
import cv2
import numpy as np
```

```python
import glob
vehicle_images=glob.glob("dataset/vehicles/*/*.png")
non_vehicle_images=glob.glob("dataset/non-vehicles/*/*.png")
print("Dataset Contains: {} Vehicle and {} Non-Vehicle images".format(len(vehicle_images),len(non_vehicle_images)))
```

Display some images

```python
num_images=5
_vIms=np.random.randint(0,len(vehicle_images),size=num_images)
_nvIms=np.random.randint(0,len(non_vehicle_images),size=num_images)
display_image=np.array([],dtype=np.uint8).reshape((64*2,0,3))
for i in range(num_images):
#     OpenCV reads images in BGR format; matplotlib needs RGB channels
    _c = cv2.cvtColor(cv2.imread(vehicle_images[_vIms[i]]),cv2.COLOR_BGR2RGB)
    _nc= cv2.cvtColor(cv2.imread(non_vehicle_images[_nvIms[i]]),cv2.COLOR_BGR2RGB)
    column=np.vstack((_c,_nc))
    display_image=np.hstack((display_image,column))

fig=plt.figure(figsize=(20,5))
plt.axis('off')
implot=plt.imshow(display_image)
```

RGB Histogram


```python
v_image=cv2.imread(vehicle_images[0])
v_rhist = np.histogram(v_image[:,:,2], bins=32, range=(0, 256))
v_ghist = np.histogram(v_image[:,:,1], bins=32, range=(0, 256))
v_bhist = np.histogram(v_image[:,:,0], bins=32, range=(0, 256))

bin_edges = rhist[1]
bin_centers = (bin_edges[1:]  + bin_edges[0:len(bin_edges)-1])/2

# Plot a figure with all three bar charts
fig = plt.figure(figsize=(25,5))
fig.suptitle('Vehicle RGB image histogram')
plt.subplot(131)
plt.bar(bin_centers, v_rhist[0])
plt.xlim(0, 256)
plt.title('R Histogram')
plt.subplot(132)
plt.bar(bin_centers, v_ghist[0])
plt.xlim(0, 256)
plt.title('G Histogram')
plt.subplot(133)
plt.bar(bin_centers, v_bhist[0])
plt.xlim(0, 256)
plt.title('B Histogram')


nv_image=cv2.imread(non_vehicle_images[0])
nv_rhist = np.histogram(nv_image[:,:,2], bins=32, range=(0, 256))
nv_ghist = np.histogram(nv_image[:,:,1], bins=32, range=(0, 256))
nv_bhist = np.histogram(nv_image[:,:,0], bins=32, range=(0, 256))
# Plot a figure with all three bar charts
fig = plt.figure(figsize=(25,5))
fig.suptitle('Non-vehicle RGB image histogram')
plt.subplot(131)
plt.bar(bin_centers, nv_rhist[0])
plt.xlim(0, 256)
plt.title('R Histogram')
plt.subplot(132)
plt.bar(bin_centers, nv_ghist[0])
plt.xlim(0, 256)
plt.title('G Histogram')
plt.subplot(133)
plt.bar(bin_centers, nv_bhist[0])
plt.xlim(0, 256)
plt.title('B Histogram')
```

HOG Features Test

```python
from skimage.feature import hog

print("HOG for RGB channels & Gray images")
for pix_per_cell in [4,8]:
    for cell_per_block in [2,4]:
        for orient in [7,8,9]:
            img=cv2.imread(vehicle_images[0])
            fig=plt.figure(figsize=(15,3))
            fig.suptitle("HOG: pix_per_cell: {} cells_per_block: {} orient: {}".format(pix_per_cell,cell_per_block,orient))
            plt.subplot(151)
            plt.title("Original Image")
            plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
            plt.subplot(152)
            _, hog_image_r = hog(img[:,:,2], orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of R Channel")
            plt.imshow(hog_image_r,cmap="gray")
            plt.subplot(153)
            _, hog_image_g = hog(img[:,:,1], orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of G Channel")
            plt.imshow(hog_image_g,cmap="gray")
            plt.subplot(154)
            _, hog_image_b = hog(img[:,:,0], orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of B Channel")
            plt.imshow(hog_image_b,cmap="gray")
            plt.subplot(155)
            _, hog_image_gg = hog(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of Grayscale Img")
            plt.imshow(hog_image_gg,cmap="gray")
```

```python
print("HOG for HSV channels")
for pix_per_cell in [4,8]:
    for cell_per_block in [2,4]:
        for orient in [7,8,9]:
            img=cv2.imread(vehicle_images[0])
            fig=plt.figure(figsize=(12,3))
            fig.suptitle("HOG: pix_per_cell: {} cells_per_block: {} orient: {}".format(pix_per_cell,cell_per_block,orient))
            plt.subplot(141)
            plt.title("Original Image")
            plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
            plt.subplot(142)
            _, hog_image_r = hog(cv2.cvtColor(img,cv2.COLOR_BGR2HSV)[:,:,0], orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of H Channel")
            plt.imshow(hog_image_r,cmap="gray")
            plt.subplot(143)
            _, hog_image_g = hog(cv2.cvtColor(img,cv2.COLOR_BGR2HSV)[:,:,1], orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of S Channel")
            plt.imshow(hog_image_g,cmap="gray")
            plt.subplot(144)
            _, hog_image_b = hog(cv2.cvtColor(img,cv2.COLOR_BGR2HSV)[:,:,2], orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=False)
            plt.title("HOG of V Channel")
            plt.imshow(hog_image_b,cmap="gray")
```
